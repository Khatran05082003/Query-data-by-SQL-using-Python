# Problem 2
List community areas with per capita income less than 11000.

```python
query='select COMMUNITY_AREA_NAME,COMMUNITY_AREA_NUMBER,PER_CAPITA_INCOME FROM Census where PER_CAPITA_INCOME<11000'
data=pd.read_sql(sql=query,con=connection)
data
```

```python
plt.bar(data['COMMUNITY_AREA_NAME'],data['COMMUNITY_AREA_NUMBER'])
plt.title('COMMUNITY_AREA_NUMBER BY COMMUNITY_AREA_NAME')
plt.xlabel('COMMUNITY_AREA_NAME')
plt.ylabel('COMMUNITY_AREA_NUMBER')
```

```python
plt.bar(data['COMMUNITY_AREA_NAME'],data['PER_CAPITA_INCOME'])
plt.title('PER_CAPITA_INCOME BY COMMUNITY_AREA_NAME')
plt.xlabel('COMMUNITY_AREA_NAME')
plt.ylabel('PER_CAPITA_INCOME')
```

# Problem 3
List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis)

```python
query="select PRIMARY_TYPE,CASE_NUMBER from Crime where PRIMARY_TYPE in ('WEAPONS VIOLATION','BATTERY','CONCEALED CARRY LICENSE VIOLATION','MOTOR VEHICLE THEFT','THEFT','SEX OFFENSE')"
data=pd.read_sql(sql=query,con=connection)
data
```

```python
sns.countplot(data=data,x='PRIMARY_TYPE')
plt.title('COUNT CASE NUMBERS FOR CRIMES INVOLVING MINORS')
```

# Problem 5

# Problem 6
List the average safety score for each type of school.

```python
query="select NAME_OF_SCHOOL,AVG(SAFETY_SCORE) as AVARAGE_OF_SAFETY_SCORE from [Public] group by NAME_OF_SCHOOL"
data=pd.read_sql(sql=query,con=connection).head()
data
```

```python
plt.bar(data['NAME_OF_SCHOOL'],data['AVARAGE_OF_SAFETY_SCORE'])
plt.title('TOP 5 AVARAGE_OF_SAFETY_SCORE BY NAME_OF_SCHOOL')
plt.xlabel('NAME_OF_SCHOOL')
plt.ylabel('AVARAGE_OF_SAFETY_SCORE')
```

# Problem 7
List 5 community areas with highest % of households below poverty line

```python
query="select TOP (5) PERCENT_HOUSEHOLDS_BELOW_POVERTY,COMMUNITY_AREA_NAME from Census order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc"
data=pd.read_sql(sql=query,con=connection)
data
```

```python
plt.bar(data['COMMUNITY_AREA_NAME'],data['PERCENT_HOUSEHOLDS_BELOW_POVERTY'])
plt.title('TOP 5 PERCENT_HOUSEHOLDS_BELOW_POVERTY BY COMMUNITY_AREA_NAME')
plt.xlabel('COMMUNITY_AREA_NAME')
plt.ylabel('PERCENT_HOUSEHOLDS_BELOW_POVERTY')
```

# Problem 8
Which community area is most crime prone?

```python
query="select COMMUNITY_AREA_NAME from [Public] where SAFETY_SCORE=(select Min(SAFETY_SCORE) from [PUBLIC])"
pd.read_sql(sql=query,con=connection)
```

# Problem 9
Use a sub-query to find the name of the community area with highest hardship index

```python
query="select COMMUNITY_AREA_NAME from Census where HARDSHIP_INDEX=(select Max(HARDSHIP_INDEX) from Census)"
pd.read_sql(sql=query,con=connection)
```

# Problem 10
Use a sub-query to determine the Community Area Name with most number of crimes?

```python
query="select TOP(1) COMMUNITY_AREA_NUMBER,count(PRIMARY_TYPE) as number from Crime group by COMMUNITY_AREA_NUMBER having COMMUNITY_AREA_NUMBER is not null order by count(PRIMARY_TYPE) desc"
pd.read_sql(sql=query,con=connection)
```
